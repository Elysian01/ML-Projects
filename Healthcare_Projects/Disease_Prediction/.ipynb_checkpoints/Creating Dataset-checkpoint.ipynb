```python
l = [0]*8
print(l)
zero_count = 1
for i in range(1, 8):
    for j in range(0, 9-i):
        if (zero_count == 1):
            l[j] = 1
        elif (zero_count == 2):
            l[j] = l[j+1] = 1
        elif (zero_count == 3):
            l[j] = l[j+1] = l[j+2] = 1
        elif (zero_count == 4):
            l[j] = l[j+1] = l[j+2] = l[j+3] = 1

        print(l)
        l = [0]*8

    zero_count += 1
    if (zero_count == 5):
        break
```
